```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("/Users/palakgupta/Downloads/Dataset - Loan Default.csv")
data.rename(columns = {'VALUE':'Property Value'}, inplace = True)
data
```

```python
data.info()
```

```python
data.describe()
```

```python
# Distribution of BAD
data['BAD'].value_counts().plot(kind='bar', color=['green', 'red'])

plt.xlabel('BAD (0=Repaid, 1=Defaulted Client)')
plt.ylabel('Count')
plt.title('Distribution of BAD (Loan Default)')
plt.show()
```

```python
# Distribution of LOAN
loan = data['LOAN']
loan_summary = loan.describe()

plt.figure(figsize=(10, 6))
sns.histplot(loan, color='green')
plt.title('Distribution of Loan Amounts')
plt.xlabel('Loan Amount')
plt.ylabel('Frequency')
plt.show()
loan_summary
```

```python
# Distribution of MORTDUE
mortgage = data['MORTDUE']
mortgage_summary = mortgage.describe()

plt.figure(figsize=(10, 6))
sns.histplot(mortgage, color='orange')
plt.title('Distribution of Mortgage Amounts')
plt.xlabel('Mortgage Amount')
plt.ylabel('Frequency')
plt.show()
mortgage_summary
```

```python
# Distribution of Property Value
Property_Value = data['Property Value']
Property_Value_summary = Property_Value.describe()

plt.figure(figsize=(10, 6))
sns.histplot(Property_Value, color='blue')
plt.title('Distribution of Property Value')
plt.xlabel('Property Value')
plt.show()
Property_Value_summary
```

```python
# Distribution of Years on Job
YOJ = data['YOJ']
YOJ_summary = Property_Value.describe()

plt.figure(figsize=(10, 6))
sns.histplot(YOJ, color='pink')
plt.title('Distribution of YOJ')
plt.xlabel('Frequency')
plt.show()
YOJ_summary
```

```python
# Distribution of CLAIGE
CLAGE = data['CLAGE']
CLAGE_summary = CLAGE.describe()

plt.figure(figsize=(10, 6))
sns.histplot(CLAGE, color='yellow')
plt.title('Distribution of CLAGE')
plt.xlabel('Frequency')
plt.show()
CLAGE_summary
```

```python
# Distribution of CLNO
CLNO = data['CLNO']
CLNO_summary = CLNO.describe()

plt.figure(figsize=(10, 6))
sns.histplot(CLNO, color='red')
plt.title('Distribution of CLNO')
plt.xlabel('Frequency')
plt.show()
CLNO_summary
```

```python
# Distribution of DEBTINC
DEBTINC = data['DEBTINC']
DEBTINC_summary = DEBTINC.describe()

plt.figure(figsize=(10, 6))
sns.histplot(DEBTINC, color='brown')
plt.title('Distribution of DEBTINC')
plt.xlabel('Frequency')
plt.show()
DEBTINC_summary
```

```python
# Distribution of DEROG
data['DEROG'].plot(kind='hist', color='orange', bins = 20)
plt.xlabel('DEROG (Number of Derogatory Reports)')
plt.ylabel('Frequency')
plt.title('Distribution of DEROG (Number of Derogatory Reports)')
```

```python
# Distribution of DELINQ
data['DELINQ'].plot(kind='hist', color='pink', bins = 50)
plt.xlabel('DELINQ (Number of delinquent credit lines )')
plt.ylabel('Frequency')
plt.title('Distribution of DELINQ(Number of delinquent credit lines )')
```

```python
# Distribution of NINQ
data['NINQ'].plot(kind='hist', color='red', bins = 40)
plt.xlabel('NINQ (Number of recent credit inquiries)')
plt.ylabel('Frequency')
plt.title('Distribution of NINQ(Number of recent credit inquiries)')
```

```python
#Distribution of Reason
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='REASON', color = 'red')

plt.xlabel('Reason for Loan Application')
plt.ylabel('Frequency')
plt.title('Distribution of Reasons for loan')

plt.show()
```

```python
#Distribution of Job
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='JOB', color = 'blue')

plt.xlabel("Applicant's Job")
plt.ylabel('Frequency')
plt.title("Distribution of Applicant's Job")

plt.show()
```

```python
# Pairplot for continuous variables
sns.pairplot(data[['LOAN', 'Property Value', 'MORTDUE', 'DEROG', 'DELINQ']])
plt.suptitle('Pairplot of Continuous Variables', fontsize=20)
plt.show()
```

```python

data.groupby(['REASON', 'JOB']).size().unstack().plot(kind='bar', stacked=True, figsize=(10,6))
plt.title('Stacked Bar Plot of REASON vs JOB')
plt.xlabel('REASON')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```
